```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('/content/shapes-02.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
ret, thresh = cv2.threshold(image_gray, 80, 255, cv2.THRESH_BINARY)

np.savetxt('thresh.txt', thresh, fmt="%d")
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(1, 3, 3)
plt.imshow(thresh, cmap='gray')
plt.title('Thresholded Image')

plt.tight_layout()
plt.show()
```

```python
T_new = 120
ret, thresh = cv2.threshold(image_gray, T_new, 255, cv2.THRESH_BINARY)
np.savetxt('thresh.txt', thresh, fmt="%d")
plt.imshow(thresh, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```

```python
his = cv2.calcHist([image_gray], [0], thresh, [256], [0, 256])
fig, ax = plt.subplots()
ax.plot(his)
ax.set_title("Grayscale Histogram")
ax.set_xlabel("Grayscale value")
ax.set_ylabel("Pixels count")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: '{:1.0f}'.format(x)))
plt.show()
```

Qua đồ thị histogram ta có thể thấy
- với t = 80 thì pixels ở mức cao sẽ cho ta ảnh không được như ảnh hơi ảo không được thật so với ảnh góc hay so với ảnh gray, và các khơảng từ 150 - 200 cũng có mức pix cao lên khả năng cao ảnh cũng có nhiều đốm trắng nhưng sẽ ít hơn so với t = 80
- vơi t mới ta chọn một khoảng pixels bình thường ổn định như từ khoảng 100 - 150 hoặc 200-255 ta sẽ thây ảnh không còn bị nhiễu ảo so với t = 80, mà ta sẽ được ảnh gần như giống với ảnh gốc và ảnh gray về các sự vật trong ảnh nhưng màu sắc sẽ khác hoàn toàn nhau
- khi ảnh t = 80 thì ảnh sẽ xuất hiện nhiều đốm trắng làm ảnh trông ảo không được đẹp so với khi ảnh t = 120(trong khoảng pix không cao) thì ta sẽ được ảnh không còn đốm trắng và hoàn chỉnh các chi tiết và màu sắc đen và trắng
