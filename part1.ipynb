```python
#cau1  Import tất cả các thư viện cần thiết
import cv2
```

```python
#2. Tải hình ảnh
image = cv2.imread('/content/sample_data/image/flowers.jpeg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
plt.show()
```

```python
#Hiển thị hình ảnh ở không gian màu RGB
def imshow(title = " ", image = None ):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
imshow("Flowes", image)
```

```python
cv2.imwrite('/content/sample_data/image/output.jpg', image)
```

```python
#4. Chuyển hình ảnh từ không gian màu RGB sang Grayscale, hiển thị hình ảnh Grayscale
img = cv2.imread('/content/sample_data/image/nemo0.jpg')
imshow("Nemo", img)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
#6. In ra chiều (dimensions) của hình ảnh RGB  và hình ảnh Grayscale.
print(img.shape)
print(gray_img.shape)
```

```python
#5. Lưu hình ảnh Grayscale với tên khác img_grayscale.png
cv2.imwrite('/content/sample_data/image/img_grayscale.jpg', gray_img)
```
