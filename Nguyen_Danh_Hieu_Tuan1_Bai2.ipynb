```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('/content/sample_data/image/flowers.jpeg')
#cau1:
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(img_rgb)
print(img_rgb.shape[:2])

zeros = np.zeros(img_rgb.shape[:2], dtype="uint8")
red = cv2.merge([R, zeros, zeros])
green = cv2.merge([ zeros, G, zeros])
blue = cv2.merge([ zeros, zeros, B])
titles = ["original", "Red","Green", "Blue"]
images = [img_rgb, red, green, blue]
count = 4

for i in range(count):
    plt.subplot(2,2, i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
plt.subplots_adjust( hspace=0.5)
plt.show()
```

```python
#cau2
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
#plt.imshow(img_hsv, cmap = 'hsv')

h, s, v = cv2.split(img_hsv)
plt.figure(figsize = (10, 3))
plt.subplot(1,3,1)
plt.imshow(h, cmap='hsv')
plt.colorbar()
plt.title("Hue")

plt.subplot(1,3,2)
plt.imshow(s, cmap='gray')
plt.colorbar()
plt.title("Saturation")

plt.subplot(1,3,3)
plt.imshow(v, cmap='gray')
plt.colorbar()
plt.title("Value")

plt.show()
```

```python
#cau3
img = cv2.imread('/content/sample_data/image/dog.jpg')
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

height, width = img.shape[:2]
quarter_height, quarter_width = height/3, width/3
m  = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
img_translation = cv2.warpAffine(img, m, (width, height))
plt.imshow(img_translation)
plt.show()
```

```python
#cau4
rotation_matrix = cv2.getRotationMatrix2D((width/3, height/3), 45, 1)
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))
plt.imshow(rotated_image)
plt.show()
```
